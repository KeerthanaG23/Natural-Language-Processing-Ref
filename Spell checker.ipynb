## Calculate the probability of a word in the train corpus

```python
def P(word, N=sum(WORDS.values())): 
    # Probability of `words` in word.
    return WORDS[word] / N if word in WORDS else 0
```

## 

```python
def known(words): 
    # The subset of `words` that appear in the dictionary of WORDS.
    return set(w for w in words if w in WORDS)
```

```python
def run_tests():
    # Test correction function
    test_correction('korrectud', 'corrected')
```

```python
def test_correction(input_word, expected_correction):
    result = correction(input_word, WORDS)
    assert result == expected_correction, f"Expected: {expected_correction}, Got: {result}"
    print(f"Test Passed for '{input_word}': {result}")
```

```python
def edits1(word):
    # All edits that are one edit away from `word`.
    letters    = 'abcdefghijklmnopqrstuvwxyz'
    splits     = [(word[:i], word[i:])    for i in range(len(word) + 1)]
    deletes    = [L + R[1:]               for L, R in splits if R]
    transposes = [L + R[1] + R[0] + R[2:] for L, R in splits if len(R) > 1]
    replaces   = [L + c + R[1:]           for L, R in splits if R for c in letters]
    inserts    = [L + c + R               for L, R in splits for c in letters]
    return set(deletes + transposes + replaces + inserts)
```

```python
def edits2(word): 
    # All edits that are two edits away from `word`.
    return (e2 for e1 in edits1(word) for e2 in edits1(e1))
```

```python
def candidates(word):
    return (known([word]) or known(edits1(word)) or known(edits2(word)) or [word])
```

```python
def correction(word, context):
    # Generate possible corrections
    candidates_list = candidates(word)

    # Calculate edit distance for each candidate
    distances = {candidate: edit_distance(word, candidate) for candidate in candidates_list}

    # Filter candidates based on probability and presence in context
    filtered_candidates = [candidate for candidate in candidates_list if candidate in context]
    if filtered_candidates:
        return min(filtered_candidates, key=lambda x: distances[x])
    else:
        return min(candidates_list, key=lambda x: distances[x])
```

```python
def spell_check_sentence(sentence):
    words = re.findall(r'\b\w+\b', sentence)
    corrected_sentence = []
    for word in words:
        corrected_word = correction(word, WORDS)
        corrected_sentence.append(corrected_word)
    return ' '.join(corrected_sentence)
```

```python
from transformers import BertTokenizer, BertForMaskedLM
import torch

# Load pre-trained BERT model and tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertForMaskedLM.from_pretrained('bert-base-uncased')
model.eval()

# Define function for spell correction and homophone differentiation
def spell_correct_with_context(sentence):
    # Tokenize the input sentence
    tokens = tokenizer.tokenize(sentence)
    indexed_tokens = tokenizer.convert_tokens_to_ids(tokens)
    
    # Mask one token in the middle of the sentence
    masked_index = len(tokens) // 2
    tokens[masked_index] = '[MASK]'
    indexed_tokens[masked_index] = tokenizer.mask_token_id
    
    # Convert tokens to PyTorch tensors
    tokens_tensor = torch.tensor([indexed_tokens])
    
    # Get predictions from the model
    with torch.no_grad():
        outputs = model(tokens_tensor)
        predictions = outputs[0]
    
    # Get the predicted token for the masked position
    predicted_index = torch.argmax(predictions[0, masked_index]).item()
    predicted_token = tokenizer.convert_ids_to_tokens([predicted_index])[0]
    
    # Replace the masked token with the predicted token
    tokens[masked_index] = predicted_token
    
    # Convert tokens back to text
    corrected_sentence = ' '.join(tokens)
    
    return corrected_sentence

# Test with example sentence
sentence = "Mr Patrick is our new principle."
corrected = spell_correct_with_context(sentence)
print("Original sentence:", sentence)
print("Corrected sentence:", corrected)
```
